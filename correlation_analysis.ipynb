```python
import pandas as pd
import os

# Check if df is already loaded
if 'df' not in locals() and 'df' not in globals():
    file_path = '/content/drive/My Drive/targets-selected.feather'
    # Check if the file exists before attempting to load
    if os.path.exists(file_path):
        df = pd.read_feather(file_path)
        print("DataFrame 'df' loaded successfully.")
    else:
        print(f"Error: Feather file not found at {file_path}")
else:
    print("DataFrame 'df' is already loaded.")
```

Use the target_columns list to select the relevant columns from the DataFrame.

Reasoning: Filter the column names based on the specified criteria (containing ':120:' and ending with ':pnlPercent') and then select these columns from the DataFrame.

```python
# Get the list of all column names
column_names = df.columns.tolist()

# Initialize an empty list for target columns
target_columns = []

# Iterate through column names and filter
for col in column_names:
    if ':120:' in col and col.endswith(':pnlPercent'):
        target_columns.append(col)

# Select the target columns from the DataFrame
df_selected = df[target_columns]

# Print the target columns
print("Target columns:")
print(target_columns)

# Display the first few rows of the df_selected DataFrame
display(df_selected.head())
```

Calculate Correlation Matrix

Subtask:
Compute the pairwise correlation between the selected target columns.

Calculate the correlation matrix for the selected columns and display it.

```python
# Calculate the correlation matrix
correlation_matrix = df_selected.corr()

# Display the correlation matrix
display(correlation_matrix)
```

Visualize Correlation Matrix
Subtask:

Create a heatmap to visualize the correlation matrix.

Generate a heatmap of the correlation matrix to visually represent the relationships between the target columns.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(12, 10))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Set the title
plt.title('Correlation Matrix of Target Columns')

# Display the plot
plt.show()
```

Based on the correlation matrix and heatmap:

Strong Positive Correlations: Observe any values close to 1.0. These indicate that as one variable increases, the other tends to increase proportionally.

Strong Negative Correlations: Observe any values close to -1.0. These indicate that as one variable increases, the other tends to decrease proportionally.

Weak or No Correlations: Values close to 0.0 suggest little to no linear relationship between the variables.
